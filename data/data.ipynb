```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data1 = pd.read_excel('data1.xls', index_col=0)
data1
```

```python
data2 = pd.read_excel('data2.xls',index_col=0)
data2
```

```python
data = pd.concat([data1,data2])
data
```

```python
# Visualizar columnas
data.columns.values
```

```python
data = data.rename(columns={'Resolución':'resolucion',
                     'Clasificación':'clasificacion',
                     'Expresión':'expresion'})
data.columns.values
```

```python
# Resetea el índice y establece un nuevo índice consecutivo
data = data.reset_index(drop=True)
data
```

```python
# Cambia el valor de index a id
data.index.name = 'id'
```

```python
def eliminar_duplicados(celda):
    if isinstance(celda, str):  # Verifica si la celda es una cadena de texto
        return ', '.join(pd.unique(celda.split(', ')))
    return celda  # Si no es una cadena, devuelve la celda original
```

```python
# Aplicar la función a la columna 'departamento'
data['Departamento'] = data['Departamento'].apply(eliminar_duplicados)
data['Provincia'] = data['Provincia'].apply(eliminar_duplicados)
data['Distrito'] = data['Distrito'].apply(eliminar_duplicados)
```

```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
engine = create_engine(os.getenv("DATABASE_URL"))
```

```python
data.to_sql(name='expresiones', con=engine, if_exists='replace', index=True)
```
